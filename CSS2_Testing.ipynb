```python
import pandas as pd 
import numpy as np 
import nltk 
import tensorflow as tf
import bert_sklearn; 


import pickle
from sklearn.metrics import accuracy_score
```

```python
print(bert_sklearn.__version__)
```

```python
df = pd.read_excel("Data.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
df.count()
```

```python
df['Impact'].value_counts()
```

```python
# Create a target column using the Orientation column

target = df['Orientation']
```

```python
# Drop the ID, PersonID and DOI columns

df = df.drop(['ID', 'DOI', 'PersonID', 'Orientation', 'Impact', 'Gender', 'Area of  research - person',
       'Area of research - publication', 'UT', 'SOURCE YEAR', 'Doc type',
       'Born', 'Position'], axis=1)
```

```python
#  Check columns

df.columns
```

```python
# Split the data into training and testing set 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3, random_state=0)

# Check the shape of the training and testing set

print(X_train.shape, y_train.shape)
```

```python
# Create a BERT model 

model = bert_sklearn.BertClassifier()

# Fit the model

model.fit(X_train, y_train)

# Predict the model 

y_pred = model.predict(X_test)

```

```python
from transformers import *
```

```python
# save the model to disk

filename = 'BERT_model.sav'

pickle.dump(model, open(filename, 'wb'))
```

```python
# test the accuracy of the model



accuracy_score(y_test, y_pred)
```
